**5.Perform meaningful visualizations using the dataset.**

```python
plt.bar(iris_data['Classification'],iris_data['PW'],color='Yellow',width=0.5)
plt.xticks(rotation=360)
plt.title('Iris dataset')
plt.xlabel('Classifications',fontsize=12)
plt.ylabel('Petal Width')
```

```python
plt.bar(iris_data['Classification'],iris_data['SW'],color='Blue',width=0.5)
plt.xticks(rotation=360)
plt.title('Iris dataset')
plt.xlabel('Classifications',fontsize=12)
plt.ylabel('Sepal Width')
```

```python
plt.bar(iris_data['Classification'],iris_data['SL'],color='Red',width=.5)
plt.bar(iris_data['Classification'],iris_data['SW'],color='Blue',width=.5)
plt.title('Iris dataset')
```

```python
plt.bar(iris_data['Classification'],iris_data['PL'],color='Yellow',width=.4)
plt.bar(iris_data['Classification'],iris_data['PW'],color='Orange',width=.4)
```

```python
# box plot
plt.boxplot(iris_data[['SL','SW','PL','PW']])
plt.title('Iris dataset')
```
